```python
import numpy as np
import pandas as pd
```

```python
# subplot 내 axes에 플롯 그리기
# plt.plot() 함수로 플롯 그리는 경우, 현재 활성화되어있는 figure의 맨 마지막 위치에 해당하는 subplot의 axes부터 차례대로 플롯이 그려진다.
plt.plot(np.random.randn(50).cumsum())
```

```python
# 특정 axes를 나타내는 변수 ax1을 직접 지정하여 플롯을 그리는 경우, 해당 axes에 플롯을 그린다.
ax1.hist(np.random.randn(100), bins=20)
```

```python
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
# fig피겨에 2X3 의 suplot 배치 , 각각의 내부에 axes를 생성 
fig, axes = plt.subplots(2,3)
```

```python
fig = plt.figure()
```

```python
ax = fig.add_subplot(1,1,1)
```

```python
ax.plot(np.random.randn(1000).cumsum())
```

```python
# 플롯의 수평축 혹은 수직축에 나타난 눈금을 matplotlib에서는 '틱(tick)'이라고 부릅니다. 특별히 수평축의 눈금은 'xtick', 수직축의 눈금은 'ytick'이라고 부릅니다. 수평축의 눈금을 다른 것으로 변경하고자 할 경우, ax.set_xticks() 함수를 사용합니다.
```

```python
# 수평축 눈금 지정
ax.set_xticks([0,250,500,750,1000])
```

```python
# 눈금 문자열로 입력. rotaion 회전각도 , fontsize = "small"
labels = ax.set_xticklabels(["one","two","three","four","five"], rotation=30, fontsize="small")
```

```python
# axis의 제목입력
ax.set_title("Random walk plot")
```

```python
ax.set_xlabel("Stages")
ax.set_ylabel("Values")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
```

```python
ax.plot(np.random.randn(1000).cumsum(), 'k', label="one")
ax.plot(np.random.randn(1000).cumsum(), "b--", label="two")
ax.plot(np.random.randn(1000).cumsum(), "r.", label="three")
```

```python
# 범례 표시하기 loc="best" 는 제시된 axes상에서 최적의 위치에 범례를 자동으로 배치
ax.legend(loc="best")
```

```python
# 수평축 값 범위 표시
ax.get_xlim()
```

```python
# 수평축 값 설정
ax.set_xlim([100,900])
```

```python
# 수직축 값 설정
ax.set_ylim([100,900])
```
